```python
# Identify year columns (starting with "YR")
year_columns = [col for col in economic_df.columns if col.startswith("YR")]

# Reshape from wide to long format (Keep only Country, Indicator, and Year)
economic_indicators_long = economic_df.melt(
    id_vars=["Country", "Indicator"],  # Fixed columns
    value_vars=year_columns,  # Pivot years
    var_name="Year",
    value_name="Value"
)

# Convert "Year" column from "YR2001" → 2001 (integer)
economic_indicators_long["Year"] = economic_indicators_long["Year"].str.extract("(\d+)").astype(int)

# Keep only the required three columns
#economic_indicators_final = economic_indicators_long.drop(columns=["value"])
```

```python
# 3. Extract Temporal Features
date_cols = ['disbursement_date', 'due_date']
for col in date_cols:
    combined_df[col] = pd.to_datetime(combined_df[col], dayfirst=True, errors='coerce')  

    # Extract temporal features with the required column names
    combined_df[f"{col}_month"] = combined_df[col].dt.month.fillna(0).astype(int)
    combined_df[f"{col}_day"] = combined_df[col].dt.day.fillna(0).astype(int)
    combined_df[f"{col}_year"] = combined_df[col].dt.year.fillna(0).astype(int)
    combined_df[f"{col}_weekday"] = combined_df[col].dt.weekday.fillna(0).astype(int)  # 0 = Monday, 6 = Sunday
    combined_df[f"{col}_quarter"] = combined_df[col].dt.quarter.fillna(0).astype(int)

# Rename extracted features to match the required names
combined_df.rename(columns={
    'disbursement_date_month': 'disbursement_month',
    'disbursement_date_day': 'disbursement_day',
    'disbursement_date_year': 'disbursement_year',
    'disbursement_date_weekday': 'disbursement_weekday',
    'disbursement_date_quarter': 'disbursement_quarter',
    
    'due_date_month': 'due_month',
    'due_date_day': 'due_day',
    'due_date_year': 'due_year',
    'due_date_weekday': 'due_weekday',
    'due_date_quarter': 'due_quarter'
}, inplace=True)
```

```python
# 4. Create Cyclical Features (Seasonality)
for col in ["disbursement_month", "due_month"]:
    combined_df[f"{col}_sin"] = np.sin(2 * np.pi * combined_df[col] / 12)
    combined_df[f"{col}_cos"] = np.cos(2 * np.pi * combined_df[col] / 12)
```

```python
# 5. Feature Engineering
combined_df["repayment_ratio"] = combined_df["Total_Amount_to_Repay"] / combined_df["Total_Amount"]
combined_df["log_Total_Amount"] = np.log1p(combined_df["Total_Amount"])
combined_df["duration_ratio"] = combined_df["Total_Amount"] / combined_df["duration"]
```

```python
# 6. Handle Missing Values (Fill with Country & Indicator-Specific Mean)
economic_indicators_final["Value"] = economic_indicators_long.groupby(["Country", "Indicator"])["Value"].transform(lambda x: x.fillna(x.mean()))
```

### Using Groupby

```python
# Sort data for proper rolling calculation
economic_indicators_final = economic_indicators_final.sort_values(by=["Country", "Indicator", "Year"])

# Function to calculate rolling averages
def calculate_rolling_avg(df, indicator_name, col_name, window):
    df[col_name] = (
        df[df["Indicator"] == indicator_name]
        .groupby("Country")["Value"]
        .transform(lambda x: x.rolling(window, min_periods=1).mean())
    )
    return df

# Compute 3-year and 5-year averages for Inflation
economic_indicators_final = calculate_rolling_avg(
    economic_indicators_final, 
    "Inflation, consumer prices (annual %)", 
    "Average_Inflation_3Yrs", 3
)

economic_indicators_final = calculate_rolling_avg(
    economic_indicators_final, 
    "Inflation, consumer prices (annual %)", 
    "Average_Inflation_5Yrs", 5
)

# Compute 3-year and 5-year averages for Interest Rates
economic_indicators_final = calculate_rolling_avg(
    economic_indicators_final, 
    "Lending interest rate (%)",  # Replace with actual name
    "Average_Interest_Rates_3Yrs", 3
)

economic_indicators_final = calculate_rolling_avg(
    economic_indicators_final, 
    "Lending interest rate (%)",  # Replace with actual name
    "Average_Interest_Rates_5Yrs", 5
)
```

```python
# Rename 'Country' to 'country_id' in economic_indicators_final
economic_indicators_final = economic_indicators_final.rename(columns={"Country": "country_id"})

# Merge economic indicators into combined dataset
combined_df = combined_df.merge(
    economic_indicators_final[["country_id", "Year", "Average_Inflation_3Yrs", "Average_Inflation_5Yrs",
                               "Average_Interest_Rates_3Yrs", "Average_Interest_Rates_5Yrs"]],
    on=["country_id", "Year"],
    how="left"
)
```
